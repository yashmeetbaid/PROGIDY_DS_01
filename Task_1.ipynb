```python
import pandas as pd

file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_900.csv'
data = pd.read_csv(file_path, skiprows=4)
data.head()
```

```python
import matplotlib.pyplot as plt

country_name = "Afghanistan"
afghanistan_data = data[data["Country Name"] == country_name].iloc[:, 4:-1].T
afghanistan_data.columns = ["Population"]
afghanistan_data.index = afghanistan_data.index.astype(int)
```

Plot 1: Population distribution by year for Afghanistan

```python
plt.figure(figsize=(12, 6))
plt.plot(afghanistan_data.index, afghanistan_data["Population"], marker='o', color='blue', label=country_name)
plt.title(f"Population Distribution by Year ({country_name})", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

**This line chart shows the growth of Afghanistan's population from 1960 to the most recent year.**

Plot 2: Comparison of populations for multiple countries in 2023

```python
year = "2023"
top_countries = data[["Country Name", year]].sort_values(by=year, ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_countries["Country Name"], top_countries[year], color='skyblue')
plt.title(f"Top 10 Countries by Population in {year}", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**This bar chart highlights the countries or regions with the highest populations in 2023.**

Plot 3: Histogram of population growth rates over a specific period

```python
start_year = "1960"
end_year = "2023"

growth_data = data[[start_year, end_year]].dropna()
growth_data["Growth Rate (%)"] = ((growth_data[end_year] - growth_data[start_year]) / growth_data[start_year]) * 100
plt.figure(figsize=(12, 6))
plt.hist(growth_data["Growth Rate (%)"], bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title(f"Histogram of Population Growth Rates ({start_year} to {end_year})", fontsize=16)
plt.xlabel("Population Growth Rate (%)", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**This histogram displays the distribution of population growth rates (%) from 1960 to 2023 across countries. Most countries have moderate growth rates, but there are a few with significantly higher growth rates (outliers).**
